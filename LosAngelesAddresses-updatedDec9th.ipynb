### We need to keep STR_SFX_CD, but do we need STR_SFX_DIR_CD? Can an address have two directions, like "424 N Baker St SE" ?

```python
#Checking the column STR_SFX_DIR_CD:
unique_values = df['STR_SFX_DIR_CD'].unique()
```

```python
print(unique_values)
```

### So, based on this, we may need to keep the column STR_SFX_DIR_CD, though I wonder whether people really use it. My daughter has lived at 4 different addresses in Los Angeles over the past 6 years and never had a "street suffix direction code" as part of her address. I am inclined to drop this column but I won't yet.

### Here is the new dataframe with the (possibly) unneeded columns dropped:

```python
df.drop(['HSE_ID', 'PIN', 'PIND', 'X_COORD_NBR', 'Y_COORD_NBR', 'ASGN_STTS_IND', 'ENG_DIST', 'CNCL_DIST'], axis=1, inplace=True)
```

```python
df.head()
```

### I want to see what is in the UNIT_RANGE column:

```python
unique_values = df['UNIT_RANGE'].unique()
print(unique_values)
```

### So these are obviously apartment or "unit" ranges. Since they are ranges, maybe we can drop this column. 
### If the user enters an address that matches one of the addresses in the dataframe, we could just allow them to enter an "apartment number" or "unit number" unchecked.

#### Dropping Unit Range:

```python
df.drop(['UNIT_RANGE'], axis=1, inplace=True)
```

### Get number of unique values for each column

```python
dict = {}
for i in list(df.columns):
    dict[i] = df[i].value_counts().shape[0]

pd.DataFrame(dict,index=["unique count"]).transpose()
```

### Observations from above:
- HSE_DIR_CD and STR_SFX_DIR_CD may contain only N, S, E, W. Will check below.
- There are way more house fraction numbers than expected. Some of these may be garbage values, will check below.
- LAT and LON being so close indicates a square-ish grid, I guess.

### Checking unique values in HSE_DIR_CD and STR_SFX_DIR_CD:

```python
unique_values = df['HSE_DIR_CD'].unique()
print(unique_values)
```

```python
unique_values = df['STR_SFX_DIR_CD'].unique()
print(unique_values)
```

### So the above are basically the same. Again, see my above comment that this "suffix direction" may not be needed at all. I have not seen them in any Los Angeles addresses. Also, there are only around 2000 non-null values in this column.

### Checking unique values in house fraction numbers:

```python
unique_values = df['HSE_FRAC_NBR'].unique()
print(unique_values)
```

### Some odd fractions, like 4/7. I wonder if something like that really just means unit/apartment number 4 out of 7.

### I think it is safe to drop STR_SFX_DIR_CD for reasons mentioned above.

```python
df.drop(['STR_SFX_DIR_CD'], axis=1, inplace=True)
```
